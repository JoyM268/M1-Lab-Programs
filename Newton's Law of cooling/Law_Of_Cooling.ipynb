```python
#Solving Newton's law of cooling by solution. The solution of mathematical representation of Newton's Law of cooling is, T = t2 + (t1 - t2)e^-kt, where, T=temperature at any time t, t1 = Initial temperature, t2 = surrounding temperature, k = thermal conductivity of the material.
#The temperature of a body drops from 100 C to 75 C in 10 minutes where the surrounding air is at the temperature 20 C. What will be the temperature of the body after half an hour? Plot the graph of cooling.
import numpy as np
from sympy import *
from matplotlib import pyplot as plt
t2 = 20 
t1 = 100
t = 10
T = 75
k1 = (1/t)*log((t1-t2)/(T-t2)) 
print('k = ',k1)
k = Symbol('k')
t = Symbol('t')
T = Function('T')(t)
T = t2+(t1-t2)*exp(-k*t)
print ('T = ',T)
T = T.subs(k,k1)
T = lambdify(t,T)
t = np.linspace(0, 70)
plt.plot(t, T(t), color = 'red')
plt.grid()
plt.show()
print('When time t = 30 minute T is, ',T(30), 'o C')
```
