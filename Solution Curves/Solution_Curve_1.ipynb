```python
# Solve dy/dt = -ky with parameter k = 0.3 and y(0) = 5
import numpy as np 
from scipy.integrate import odeint
import matplotlib.pyplot as plt
def model (y,t): 
    k = 0.3
    return -k*y
y0 = 5
t = np.linspace(0,20)
y = odeint(model,y0,t)
plt.plot(t,y)
plt.title('Solution of dy/dt = - ky ; k = 0.3 , y(0) = 5')
plt.xlabel('time')
plt.ylabel ('y(t)')
plt.show()
```
